```
# librerias necesarias para el análisis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Descarga de archivo
dfbruto = pd.read_excel('https://www.dipres.gob.cl/598/articles-45401_doc_xls_GCT2.xlsx')

# Extracción de filas que contienen la información de gasto
dfbruto1 = dfbruto.iloc[17:23]
dfbruto2 = dfbruto.iloc[28:31]
dfgasto = dfbruto1.append(dfbruto2)
```

```

dfgasto.columns = ['Item', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000',
                 '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
                 '2013', '2014', '2015', '2016', '2017', '2018', '2019'] 
```

```
dfgasto_t = pd.melt(dfgasto, id_vars=['Item'], value_vars=['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000',
                 '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
                 '2013', '2014', '2015', '2016', '2017', '2018', '2019'])
```

```
plt.figure(figsize=(16,8))
ax = sns.lineplot(x='variable', y='value', data=dfgasto_t, hue='Item')
ax.set_ylabel('Gasto en millones de pesos 2019')
ax.set_xlabel('Año presupuestario')
ax.set_title('Evolución anual de gasto gobierno central por Item de Estado de Operaciones, millones de pesos 2019')
```

```
dfgasto_p = dfgasto.set_index('Item')
dfgasto_p = dfgasto_p.divide(dfgasto_p.sum(), axis=1)
```

```
plt.figure(figsize=(10,6))
ax = sns.heatmap(dfgasto_p,cmap="Blues")
plt.show()
```
